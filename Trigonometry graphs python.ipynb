```python
from matplotlib import pyplot as plt
#from matplotlib import style
import numpy as np
import math

#style.use('background')

x=np.arange(0,math.pi*2,0.05)
y=np.sin(x)
plt.plot(x,y)

plt.xlabel("angle")
plt.ylabel("sine")
plt.title("sinewave")
plt.show()
```

```python
x=np.arange(0,math.pi*2,0.05)
y=np.cos(x)
plt.plot(x,y)

plt.xlabel("angle")
plt.ylabel("cos")
plt.title("coswave")
plt.show()
```

```python
x=np.arange(0,math.pi*2,0.05)
y=np.tan(x)
plt.plot(x,y)

plt.xlabel("angle")
plt.ylabel("tan")
plt.title("tanwave")
plt.show()
```
